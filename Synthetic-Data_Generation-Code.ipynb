```python
from faker import Faker
import pandas as pd
import random

# Initialize Faker
faker = Faker()

# Define predefined lists for skills and degrees
degrees = ["B.Tech", "M.Tech", "MBA", "B.Sc", "M.Sc", "PhD", "Diploma"]
skills_list = [
    "Python", "Java", "Machine Learning", "Deep Learning", "SQL", 
    "Data Analysis", "Cloud Computing", "Web Development", 
    "Cybersecurity", "AI", "Project Management"
]

# Generate synthetic data
data = []
for _ in range(100):  # Adjust the range to create more records
    degree = random.choice(degrees)
    institution = faker.company()
    graduation_year = faker.random_int(min=2000, max=2024)
    experience_years = faker.random_int(min=0, max=20)
    num_skills = random.randint(3, 6)  # Number of skills for each individual
    skills = random.sample(skills_list, num_skills)
    data.append({
        "Degree": degree,
        "Institution": institution,
        "Graduation Year": graduation_year,
        "Experience (Years)": experience_years,
        "Skills": ", ".join(skills)
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("career_guidance_dataset.csv", index=False)

# Display the first few rows
print(df.head())
```

```python
from faker import Faker
import pandas as pd
import random

# Initialize Faker
faker = Faker()

# Predefined lists for degrees and skills
degrees = ["B.Tech", "M.Tech", "MBA", "B.Sc", "M.Sc", "PhD", "Diploma"]
skills_list = [
    "Python", "Java", "Machine Learning", "Deep Learning", "SQL", 
    "Data Analysis", "Cloud Computing", "Web Development", 
    "Cybersecurity", "AI", "Project Management", "C++", 
    "DevOps", "Mobile Development", "Blockchain"
]

# Generate synthetic data
data = []
for _ in range(10000):  # Adjust to generate more rows
    degree = random.choice(degrees)
    institution = faker.company()
    graduation_year = faker.random_int(min=2000, max=2024)
    num_companies = random.randint(1, 5)  # Number of companies in experience
    companies = [faker.company() for _ in range(num_companies)]
    experience = "; ".join(companies)  # Concatenate company names as experience
    num_skills = random.randint(4, 7)  # Number of skills for each individual
    skills = random.sample(skills_list, num_skills)
    data.append({
        "Degree": degree,
        "Institution": institution,
        "Graduation Year": graduation_year,
        "Experience (Companies)": experience,
        "Skills": ", ".join(skills)
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("career_guidance_dataset_large.csv", index=False)

# Display the first few rows
print(df.head())
```
